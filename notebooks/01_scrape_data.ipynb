```python
shifts = shifts_df.copy()
shifts["start_time"] = shifts["start_time"].apply(time_to_seconds)
shifts["end_time"] = shifts["end_time"].apply(time_to_seconds)

# Convert remaining time to elapsed time
shifts['start_elapsed'] = 1200 - shifts['start_time']
shifts['end_elapsed'] = 1200 - shifts['end_time']

# Convert to total game seconds
shifts['start_seconds'] = (shifts['period'] - 1) * 1200 + shifts['start_elapsed']
shifts['end_seconds'] = (shifts['period'] - 1) * 1200 + shifts['end_elapsed']
```

```python
shifts
```

```python
def get_on_ice_players(shot_row, shifts_df):
    time = shot_row['game_seconds']
    home_away = shot_row['home']

    players_on_ice = shifts_df[
        (shifts_df['start_seconds'] <= time) &
        (shifts_df['end_seconds'] >= time)
    ]

    home_team_id = shifts_df['home_id'].iloc[0]
    away_team_id = shifts_df['away_id'].iloc[0]

    home_players = players_on_ice[players_on_ice['shift_team'] == home_team_id]['player_id'].tolist()
    away_players = players_on_ice[players_on_ice['shift_team'] == away_team_id]['player_id'].tolist()

    return pd.Series({'home_players': home_players, 'away_players': away_players})
```

```python
xG_time[['home_players', 'away_players']] = xG_time.apply(
    lambda row: get_on_ice_players(row, shifts), axis=1
)
```

```python
xG_time
```

```python
from collections import defaultdict
xG_dict = {
            "xG_for": defaultdict(float),
            "xG_against": defaultdict(float)
           }

for _, row in xG_time.iterrows():
    xg = row['xG']
    is_home_shot = row['home']

    # Players on the shooting team get xG_for
    if is_home_shot == "Home":
        shooting_team_players = row['home_players']
        defending_team_players = row['away_players']
    else:
        shooting_team_players = row['away_players']
        defending_team_players = row['home_players']

    for pid in shooting_team_players:
        xG_dict["xG_for"][pid] += xg
    for pid in defending_team_players:
        xG_dict["xG_against"][pid] += xg
```

```python
xG_dict['xG_against']
```

```python
# Ensure every player is in both dictionaries
all_players = set(xG_dict["xG_for"].keys()) | set(xG_dict["xG_against"].keys())

player_xg_summary = pd.DataFrame([
    {
        'player_id': pid,
        'xG_for': xG_dict["xG_for"].get(pid, 0.0),
        'xG_against': xG_dict["xG_against"].get(pid, 0.0)
    }
    for pid in all_players
])
```
